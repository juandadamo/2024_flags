# ensayo con papel

```python
# Datos del Papel, 
Ym = 85
t = 100
dens_sup = 80
rho_papel = dens_sup/(t*1e-6)*1e-3
Papel_80 = material('Papel 80gr/m2',t,rho_papel,Ym)
B = Papel_80.B
Papel_80.rho = rho_papel
k1L = BnL[0]
k2L = BnL[1]
L = 150e-3
L = 50e-3
Papel_80.L = L
Papel_80.freq_nat()
display(Latex(f'${{f_n}}_1= {Papel_80.fn[0]:.3f}$Hz'))
display(Latex(f'${{f_n}}_2= {Papel_80.fn[1]:.3f}$Hz'))
```

```python
Papel_80.B*1e5
```

```python
# salida de datos en un dataframe

dict_papel = {'$\\rho$':Papel_80.rho,'$\\mu$':Papel_80.mu,'B=EI':Papel_80.B*1e5
             ,'$L$':Papel_80.L*1e3,'th':Papel_80.thickness*1e6,'$f_{n1}$':
              Papel_80.fn[0],'$f_{n2}$':Papel_80.fn[1]}
pd_papel = pd.DataFrame(dict_papel.items(),columns=['Propiedad','Valor'])
pd_papel['unid.']=np.zeros_like((len(pd_papel)),dtype='<U10')
pd_papel.iloc[:,2]=['kg/$m^3$','kg/m$^2$','N$\\cdot$ m$10^{-5}$','mm',
                    '$\\mu m$','Hz','Hz']

blankIndex=[''] * len(pd_papel)
pd_papel.index=blankIndex
display(pd_papel)
```
