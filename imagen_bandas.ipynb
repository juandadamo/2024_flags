```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, interact_manual,interactive,widgets,Layout
from IPython.display import Latex
import scipy as sc
import sympy as sp
import pandas as pd
from funciones_flag import *
from matplotlib.colors import LinearSegmentedColormap
colors = [(0, 0, 0), (0,1, 0)] # first color is black, last is red
cm = LinearSegmentedColormap.from_list(
        "Custom", colors, N=20)
mat = np.indices((10,10))[1]
```

```python
frecs = np.array([12,14,15,16,17,18,20,24,28])
presiones = np.array([1.7,3.3,4.5,6.1,7.3,9.9,15.3,29.6,47.6])
```

```python
rhoa = 1.2
g = 9.8
```

```python
Veloc = np.sqrt(2*presiones/rhoa)
```

```python
Veloc
```

```python
p1 = np.polyfit(frecs,Veloc,1)
f_vel = np.poly1d(p1)
```

```python
fig,ax = plt.subplots(figsize=(4,4))
lin, = ax.plot(frecs,Veloc,'o')
frecs_ = np.linspace(5,30,100)
ax.plot(frecs_,f_vel(frecs_),color = lin.get_color())
ax.grid()
ax.set_xlabel('$frec(Hz)$')
ax.set_ylabel('$U$ [m/s]');
```
