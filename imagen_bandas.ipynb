```python
frecs = np.array([10,12,14,16,18,20,22,24])
presiones = np.array([17.2,26.5, 39.5 , 54.5 , 73 , 93,115,141])
velocs = np.array([5.3, 6.7 , 8.1, 9.5,11 ,12.4,13.8,15.3])
```

```python
fig,ax = plt.subplots()
ax.plot(frecs,presiones,'sk')
ax.grid()
```

```python
rhoa = 1.2
g = 9.8


pbariloche = 91.92e3
patm = 101.3e3

pratio = pbariloche/patm

pratio
rhoa_b = rhoa*pratio
rhoa_b
```

```python
Veloc = np.sqrt(2*presiones/rhoa_b)
```

```python
p1 = np.polyfit(frecs,Veloc,1)
f_vel = np.poly1d(p1)
```

```python
fig,ax = plt.subplots(figsize=(4,4))
lin, = ax.plot(frecs,Veloc,'o')
frecs_ = np.linspace(5,30,100)
ax.plot(frecs_,f_vel(frecs_),color = lin.get_color())
ax.grid()
ax.set_xlabel('$frec(Hz)$')
ax.set_ylabel('$U$ [m/s]');
#ax.plot(frecs,velocs,'s',linestyle = 'none',fillstyle='none')
```

```python

r=5
npoints = 1500
img_black = np.zeros((npoints,npoints))
xr_i = np.linspace(4*r,npoints-4*r,int(npoints/(r*6)))
yr_i = np.linspace(4*r,npoints-4*r,int(npoints/(r*6)))
Xr_i,Yr_i = np.meshgrid(xr_i,yr_i)

amp_rand = 3
Rand_x = np.random.randn(Xr_i.shape[0],Xr_i.shape[1])*amp_rand
Rand_y = np.random.randn(Yr_i.shape[0],Yr_i.shape[1])*amp_rand
Xr = Xr_i + Rand_x
Yr = Yr_i +Rand_y
```

```python
fig,ax = plt.subplots(frameon=False)
ax.imshow(img_black)
ax.plot(Xr,Yr,marker='o',color='y',markersize=1,linestyle='none');
ax.axis('off')
fig.savefig('discos1.png')
```
